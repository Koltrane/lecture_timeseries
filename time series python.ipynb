Let's put it together

```python
def format_datetime(data):
    melted = pd.melt(data, id_vars='date')
    melted['variable'] = [timedelta(hours=int(x[1])) for x in melted.variable]
    index = melted.ix[:, ['date', 'variable']].sum(axis=1)
    return Series(melted.value.values, index).sort_index()
                     
rs = format_datetime(data)
rs
```

### Moving windows

```python
pd.rolling_mean(rets, 60, min_periods=15).between_time('10:30', '16:30')
```

```python
pd.expanding_std(rets).between_time('10:30', '16:00')
```

```python
pd.rolling_apply(rets, 60, lambda x: x.sum(), min_periods=15).between_time('10:30', '16:00')
```

###Fitting using Statsmodels

```python
from pandas.io.data import get_data_yahoo
```

```python
y = get_data_yahoo('AAPL', '2005-01-01', '2012-03-18')
```

```python
y = y.resample('W', how='first')
```

```python
rets = (np.log(y['Adj Close']) - np.log(y['Adj Close'].shift(1))).dropna()
```

```python
rets.plot()
```

```python
sm.graphics.tsa.plot_pacf(rets, lags=10)
```

```python
model = sm.tsa.ARMA(rets.dropna(), freq='D', order=(0,3))
```

```python
results = model.fit(trend='c', disp=-1)
```

```python
results.summary()
```
